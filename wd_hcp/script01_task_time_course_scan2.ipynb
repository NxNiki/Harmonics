```python
## select time points for each condition of task fMRI data.

import pandas as pd
import numpy as np
import glob
import os
import seaborn as sns
import matplotlib.pyplot as plt
import re

time_label_key = pd.read_csv('/home/xin/Downloads/HCP_Data/FullData_Oct26/timeLabels_KEY.csv', header = None, index_col = 0)
time_label_key = time_label_key.to_dict()
time_label_key = time_label_key[1]
del time_label_key['cue']
time_label_key
```

```python

input_dir = '/home/xin/Downloads/HCP_Data/AROMA_fullData_July2021/Scan2'
output_dir = 'hcp_out01_task_time_course_scan2'

files = glob.glob(input_dir + '/TimeSeries*.csv')
print(len(files))

if not os.path.isdir(output_dir):
    os.mkdir(output_dir)

for f in files:
    print(f)

    subject_id = re.search('(.*)TimeSeries(.*).csv', f).group(2)
    roi_signals = pd.read_csv(f, sep=',', header = None).T
    #print(roi_signals.shape)
    
    time_label = pd.read_csv('/home/xin/Downloads/HCP_Data/FullData_Oct26/Scan2/timingLabels_' + subject_id + '.csv', header = None)
    file_name = f.split('/')
    
    for condition in time_label_key:
        out_fname = output_dir + '/' + file_name[-1].replace('TimeSeries', 'TimeSeries_' + condition.replace(' ', '_') + '_')
        cond_index = time_label == time_label_key[condition]
        cond_signals = roi_signals.iloc[cond_index.values, :]
        cond_signals.to_csv(out_fname, index = False, columns = None)
    
    # break
```
