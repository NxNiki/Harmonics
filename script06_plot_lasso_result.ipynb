```python
## plot result of lasso:

import matplotlib.pyplot as plt
import seaborn as sns
import glob
import pandas as pd
import re
import os

# harmonics feature:
# input_dir = 'adni_out05_lasso/'
# output_dir = 'adni_out06_lasso_plot/'
# x_label = 'frequency'

# network feature:
input_dir = 'adni_out05_lasso_network_metrics/'
output_dir = 'adni_out06_lasso_plot_network_metrics/'
x_label = 'metrics'
```

```python
## plot accuracy
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
    
files = glob.glob(input_dir + 'lasso_accuracy_*.csv')

for f in files:
    
    data = pd.read_csv(f)
    group1 = re.search('lasso_accuracy_(.*)_(.*).csv', f).group(1)
    group2 = re.search('lasso_accuracy_(.*)_(.*).csv', f).group(2)
    
    data.rename(columns = {'Unnamed: 0': x_label}, inplace = True)
    
    data = pd.melt(data, id_vars = [x_label])
    data.rename(columns = {'value': 'accuracy'}, inplace = True)
    
    # sns.boxplot(data = data, x = x_label, y = 'accuracy')
    sns.swarmplot(data = data, x = x_label, y = 'accuracy')

    plt.title(group1 + ' vs ' + group2)
    plt.savefig(f.replace('.csv', '.png').replace(input_dir, output_dir))
    plt.clf() #
    
    # break
# data
```
