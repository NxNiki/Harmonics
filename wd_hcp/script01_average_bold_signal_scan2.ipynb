```python
## compute average bold signal for each condition of task fMRI data.
## the result will be used to tune the parameters for harmonic analysis.
## Xin June-14th-2022.

import pandas as pd
import numpy as np
import glob
import os
import seaborn as sns
import matplotlib.pyplot as plt
import re

time_label_key = pd.read_csv('/home/xin/Downloads/HCP_Data/FullData_Oct26/timeLabels_KEY.csv', header = None, index_col = 0)
time_label_key = time_label_key.to_dict()
time_label_key = time_label_key[1]
del time_label_key['cue']

time_label_key
```

```python

time_label_key = {'0back': [4, 9, 11, 15],
                  '2back': [2, 7, 13, 17]}
```

```python

input_dir = '/home/xin/Downloads/HCP_Data/AROMA_fullData_July2021/Scan2'
output_dir = 'hcp_out01_average_bold_signal_scan2'

files = glob.glob(input_dir + '/TimeSeries*.csv')
print(len(files))

result_list = []

for f in files:
    print(f)

    subject_id = re.search('(.*)TimeSeries(.*).csv', f).group(2)
    roi_signals = pd.read_csv(f, sep = ',', header = None).T
    #print(roi_signals.shape)
    
    time_label = pd.read_csv('/home/xin/Downloads/HCP_Data/FullData_Oct26/Scan2/timingLabels_' + subject_id + '.csv', header = None)
    file_name = f.split('/')
    
    for condition in time_label_key:
        
        output_dir2 = output_dir + '_' + condition.replace(' ', '_') + '/'
        #cond_index = time_label == time_label_key[condition]
        cond_index = time_label.isin(time_label_key[condition])
        cond_signals = roi_signals.iloc[cond_index.values, :].mean(axis = 0)
        result_list.append(cond_signals.T)
        
cond_signals_0back = pd.concat(result_list[0::2], axis = 1).transpose()
cond_signals_2back = pd.concat(result_list[1::2], axis = 1).transpose()
```

```python
if not os.path.isdir(output_dir):
    os.mkdir(output_dir)

cond_signals_0back.to_csv(output_dir + '/mean_bold_signal_0back.csv', index = False, columns = None)
cond_signals_2back.to_csv(output_dir + '/mean_bold_signal_2back.csv', index = False, columns = None)
```
