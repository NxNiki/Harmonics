```python
## use all time points on scan1 to construct correlation matrix.

import pandas as pd
import numpy as np
import glob

import os
import seaborn as sns
import matplotlib.pyplot as plt
# import re
```

```python

# for 0back and 2 back condition:
# input_dir = 'hcp_out01_task_time_course_scan2_0back'
# output_dir = 'hcp_out02_corr_matrix_scan2_0back'

input_dir = 'hcp_out01_task_time_course_scan2_2back'
output_dir = 'hcp_out02_corr_matrix_scan2_2back'

files = glob.glob(input_dir + '/TimeSeries*.csv')
print(len(files))

if not os.path.isdir(output_dir):
    os.mkdir(output_dir)

for f in files:
    #print(f)

    #roi_signals = pd.read_csv(f, sep=',', header = None).T
    
    # this signals has been tranposed in the previous step:
    roi_signals = pd.read_csv(f, sep=',', header = None)
    #print(roi_signals.shape)
    
    corr_mat = roi_signals.corr(method = 'pearson', min_periods = 100)
    #print(corr_mat)
    
    file_name = f.split('/')
    out_fname = output_dir + '/' + file_name[-1].replace('TimeSeries', 'corr_matrix_')
    
   # check if 0 columns or NA values exists:
    if corr_mat.isna().any().any():
        print('NaN values found in correlation matrix, skip.')
        print(out_fname)
        continue

    corr_mat.to_csv(out_fname, index = False, columns = None)
    
    fig, ax = plt.subplots()
    ax = sns.heatmap(corr_mat)
    fig_name = out_fname.replace('.csv', '.png')
    plt.savefig(fig_name)
    plt.close('all')
    
```
